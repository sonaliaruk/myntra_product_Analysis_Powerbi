# Data Wrangling

---> Drop unncessary columns

```python
# display all columns
df.columns
```

```python
# drop column url
df.drop(['URL'],axis = 1, inplace = True)
```

```python
df.columns
```

---> Converting wrong data types to appropriate formats

```python
df.info()
```

```python
# convert data type of discountoffer column
# First convert all values to string just in case some are numeric
df['DiscountOffer'] = df['DiscountOffer'].astype(str)

# Remove unwanted text
df['DiscountOffer'] = df['DiscountOffer'].str.replace('%', '', regex=False)
df['DiscountOffer'] = df['DiscountOffer'].str.replace('off', '', case=False, regex=True)
df['DiscountOffer'] = df['DiscountOffer'].str.replace(' ', '', regex=False)


# Convert cleaned strings to float
df['DiscountOffer'] = pd.to_numeric(df['DiscountOffer'], errors='coerce')
```

```python
# convert data type of Reviews  column
# convert from float to int
df['Reviews'] = df['Reviews'].astype('Int64')
```

---> Handling null or missing values

```python
df.isnull().sum()
```

```python
# fill rating column with 0  as no rating
df['Ratings'] = df['Ratings'].fillna(0)

# fill  review column with 0 as no review
df['Reviews'] = df['Reviews'].fillna(0)
```

```python
# Fill missing DiscountOffer using DiscountPrice (in Rs) and OriginalPrice (in Rs)
mask_offer_missing = df['DiscountOffer'].isnull() & df['DiscountPrice (in Rs)'].notnull() # filter null from DiscountOffer and non null from DiscountPrice
df.loc[mask_offer_missing, 'DiscountOffer'] = (df.loc[mask_offer_missing, 'DiscountPrice (in Rs)'] / df.loc[mask_offer_missing, 'OriginalPrice (in Rs)'] * 100)


print(df['DiscountOffer'].isnull().sum()) #Missing values reduced from 97125 to 80933 after cleaning


```

```python
# Fill missing DiscountPrice (in Rs) using DiscountOffer  and OriginalPrice (in Rs)
mask_offer_missing = df['DiscountPrice (in Rs)'].isnull() & df['DiscountOffer'].notnull() # filter null from DiscountPrice and non null from DiscountOffer
df.loc[mask_offer_missing, 'DiscountPrice (in Rs)'] = (df.loc[mask_offer_missing, 'OriginalPrice (in Rs)'] * df.loc[mask_offer_missing, 'DiscountOffer']/ 100)

df['DiscountPrice (in Rs)'].isnull().sum() #  Missing values reduced from 193158 to 80933 after cleaning
```

```python
df['DiscountPrice (in Rs)'] = df['DiscountPrice (in Rs)'].fillna(df['OriginalPrice (in Rs)'])
df['DiscountOffer'] = df['DiscountOffer'].fillna(0)
```

```python
df['OriginalPrice (in Rs)'] = df['OriginalPrice (in Rs)'].astype(int)

df['DiscountPrice (in Rs)'] = df['DiscountPrice (in Rs)'].astype(int)

df['DiscountOffer'] = df['DiscountOffer'].astype(int)
```

```python
# check outlier using boxplot
sns.boxplot(data = df , y = 'OriginalPrice (in Rs)')
plt.figure(figsize = (10,6))
plt.show()
```

```python
sns.distplot(df['OriginalPrice (in Rs)'])
plt.figure(figsize = (10,6))
plt.show()
```

```python
# Remove outliers using IQR
Q1 = df['OriginalPrice (in Rs)'].quantile(0.25)
Q3 = df['OriginalPrice (in Rs)'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['OriginalPrice (in Rs)'] >= lower) & (df['OriginalPrice (in Rs)'] <= upper)]
```
